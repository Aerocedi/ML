```python
import sklearn
import math
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.linear_model import LinearRegression
lr=LinearRegression(n_jobs=1,normalize=False)
raw_data=pd.read_csv("Advertising.csv")
print(raw_data.head())
print(raw_data.tail())
modified_data=raw_data.drop("Unnamed: 0",axis=1)
print(modified_data)
features=modified_data[['Radio','Newspaper']]
print(features)
target=modified_data['Sales']
lr.fit(features,target)
X_test=[[4.1,5.8]]
predicted=lr.predict(X_test)
print(predicted)
check=target[190]
accuracy=abs(check-predicted)
print(100-accuracy)
plt.scatter(modified_data['TV'],target, color="blue")
plt.scatter(modified_data['Radio'],target,color="brown")
plt.scatter(modified_data['Newspaper'],target,color="green")
plt.xlabel("Modified_Datas",color="green")
plt.ylabel("Sales",color="red")
plt.show()
```
